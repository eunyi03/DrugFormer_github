```python
import pickle

# Gdata.pickle 파일 열기
with open('dictionary_genename_token_pair.pickle', 'rb') as file:
    data = pickle.load(file)

# 데이터 전체 출력
print("전체 데이터 출력:")
print(data)

# 데이터의 전체 길이 출력
print(f"데이터 길이: {len(data)}")

# 데이터의 일부 출력 (예: 처음 10개 항목만)
print("데이터 일부 출력:")
for key, value in list(data.items())[:15]:  # 처음 10개 항목만 출력
    print(f"{key}: {value}")

# ID 값의 범위 확인
values = list(data.values())
min_value = min(values)
max_value = max(values)
print(f"ID 값의 최소값: {min_value}")
print(f"ID 값의 최대값: {max_value}")

# 유전자 이름의 갯수 확인
num_genes = len(data)
print(f"총 유전자 및 특수 토큰의 갯수: {num_genes}")
```

```python
# 파일 경로 설정
file_path = '/data/project/eunyi/DrugFormer/gene_summary.txt'

# 파일을 열어서 내용 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.readlines()

# 첫 5줄 출력
for line in content[:5]:
    print(line.strip())  # strip()을 사용하여 줄 끝의 공백 문자 제거
```

```python
# 첫 번째 줄을 보기 좋게 줄바꿈 추가해서 출력
for line in content[:1]:
    print(line.strip().replace('\t', '\n'))  # 탭 구분을 줄바꿈으로 변경
```
