```python
# 유전자 이름을 정규화하여 하이픈(-)과 점(.)을 동일하게 처리
def normalize_gene_name_v2(gene_name):
    return re.sub(r'\.', '-', gene_name)  # '.'을 '-'로 바꾸기

# <와 >로 시작하거나 끝나는 토큰을 제외하고, 'A', 'G', 'C', 'T' 같은 염기서열 문자를 제외하는 함수
def filter_gene_names(gene_name):
    # <와 >로 시작하거나 끝나는 토큰은 유전자 이름이 아니므로 제외
    if gene_name.startswith('<') or gene_name.endswith('>'):
        return False
    # 'A', 'G', 'C', 'T'는 유전자 이름이 아니므로 제외
    if gene_name in ['A', 'G', 'C', 'T']:
        return False
    return True

# 데이터프레임과 토큰 딕셔너리에서 유전자 이름을 정규화
genes_in_scdata_normalized_v2 = [normalize_gene_name_v2(gene) for gene in genes_in_scdata if filter_gene_names(gene)]
genes_in_token_dict_normalized_v2 = {normalize_gene_name_v2(gene): token for gene, token in dictionary_genename_token_pair.items() if filter_gene_names(gene)}

# 피클 파일에 있는 유전자들 중 scdata에 없는 유전자들 확인
missing_in_scdata = [gene for gene in genes_in_token_dict_normalized_v2 if gene not in genes_in_scdata_normalized_v2]

# scdata에만 있는 유전자들 확인
genes_in_scdata_only = [gene for gene in genes_in_scdata_normalized_v2 if gene not in genes_in_token_dict_normalized_v2]

# 결과 출력
print(f"dictionary_genename_token_pair에 있지만 scdata에는 없는 유전자 수: {len(missing_in_scdata)}")
print(f"dictionary_genename_token_pair에 있지만 scdata에 없는 유전자 목록: {missing_in_scdata}")

print(f"scdata에만 있는 유전자 수: {len(genes_in_scdata_only)}")
print(f"scdata에만 있는 유전자 목록: {genes_in_scdata_only}")

# scdata에 있는 유전자 수와 피클 파일에 있는 유전자 수 출력
print(f"scdata에 있는 유전자 수: {len(genes_in_scdata_normalized_v2)}")
print(f"dictionary_genename_token_pair에 있는 유전자 수: {len(genes_in_token_dict_normalized_v2)}")
```
